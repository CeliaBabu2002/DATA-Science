```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset= pd.read_csv("Salary_Data.csv")
dataset.head(4)
```

```python
dataset.isnull().sum()
```

```python
dataset["Salary"].fillna(dataset["Salary"].mode()[0],inplace=True)
dataset["Experience"].fillna(dataset["Experience"].mode()[0],inplace=True)
dataset["Age"].fillna(dataset["Age"].mode()[0],inplace=True)
```

```python
plt.figure(figsize=(5,4))
sns.pairplot(data=dataset)
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor,plot_tree
```

```python
dt=DecisionTreeRegressor(max_depth=5)
dt.fit(x_train,y_train)
```

```python
# there is much difference between train and test score so we added max_depth to avoid over fitting
dt.score(x_test,y_test)*100,dt.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(20,20))
plot_tree(dt)
plt.savefig("Decision_Tree_Regression.jpg")
plt.show()
```
