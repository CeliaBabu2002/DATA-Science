# Logistic Regression 
Logistic Regression is one of the mostpopular machine Learning Algorithm,which comes under the supervised Learning Technique

- Its is used to predict Categorical dependent variable using a given set of independent variable
- Therefore the outcome must be in form of category or discrete form.It can be either 0 or 1,Yes Or No,True or False,But instead of giving exact 0 as output it gives probability of values between 0 to 1

# Types of Logistic Regression
- Binomial:there can only be two types of dependent variable ie output ,like yes or no ,0 or 1

- Multinomial:there can be 3 or more types of unordered dependent variable, "cat,"dog,"cow"

- Ordinal:there can be 3 or more types of ordered dependent variable,size chart S,M,L,XL

# Logistic Regression using Single input

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#here we a creating a model that can predict purchase based on Age ,so we need to drop EstimatedSalary
dataset= pd.read_csv("Social_Network_Ads.csv")
dataset.head(4)
dataset.drop(columns=["EstimatedSalary"],inplace=True)
dataset
```

```python
plt.figure(figsize=(5,3))
plt.scatter(dataset["Age"],dataset["Purchased"])
plt.show()
```

```python
x=dataset[["Age"]]
y=dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
# finding te predict line

plt.figure(figsize=(5,3))
plt.scatter(dataset["Age"],dataset["Purchased"])
sns.lineplot(x="Age",y=lr.predict(x),data=dataset,c="red")
plt.show()
```
