# Types of Navie BAYES

# 1. Multinomial Naive Bayes
- if i/p data contains test data and if it is descrete that we us MultiNomial 
# 2. Gaussian Naive Bayes
- If the i/p data is normally distributed the we use Gaussian Model
# 3. Bernoulli Naive Bayes
- if the i/p data is in binary ie if it contain boolean data 0 or 1 then we use Bernoulli Model

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("Placement_Score.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa",y="score",data=dataset,hue="placed")
plt.legend(loc=1)
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["placed"]
```

```python
# here we can decide the model based on the data,ie
# if the data is normal distributed then we use gaussian model
# if the data is binary ie contains only 0 or 1 then bernouill model
# if the data contains text data and if it is discrete the multinomial

#here first we check if if it is normally distributed here we have to find separately of each column (feature)

#sns.kdeplot(data=dataset["cgpa"]) #score,placed
#plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB,BernoulliNB,GaussianNB
```

```python
mn=MultinomialNB()
mn.fit(x_train,y_train)
```

```python
mn.score(x_test,y_test)*100,mn.score(x_train,y_train)*100
```

```python
mn.predict([[8.27,8.86]]) #8.27	8.86	1
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mn)
plt.show()
```

```python
br=BernoulliNB()
br.fit(x_train,y_train)
```

```python
br.score(x_test,y_test)*100,br.score(x_train,y_train)*100
```

```python
br.predict([[8.14,6.52]]) #8.14	6.52	1
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=br)
plt.show()
```

```python
ga=GaussianNB()
ga.fit(x_train,y_train)
```

```python
ga.score(x_test,y_test)*100,ga.score(x_train,y_train)*100

# more accuracy since the data is normally distributed
```

```python
ga.predict([[8.27,8.86]]) #8.27	8.86	1
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=ga)
plt.show()
```
