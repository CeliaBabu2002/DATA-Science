# Polynomial Regression

Polynomial regression is a regression algorithm that amodels the relationship between a dependent(y) and independent variable(x) as nth degree polynomial

Y=b0+b1x1+b2x2^2....bnxn^n

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("polynomial.csv")
dataset.head(3)
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset["Salary"].fillna(dataset["Salary"].mode()[0],inplace=True)
```

```python
x=dataset[["Level"]]
y=dataset["Salary"]
```

```python
dataset.corr()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf= PolynomialFeatures(degree=2)
```

```python
pf.fit(x)
```

```python
x =pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
y=lr.predict(x)
```

```python
plt.scatter(dataset["Level"],dataset["Salary"])
plt.plot(dataset["Level"],y,c="red")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()

# issue with the dataset 
```
